```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x, k, x0):
    return 1 / (1 + np.exp(-k * (x - x0)))

x = np.linspace(-10, 10, 100)

midpoint_values = [-5, -2, -1, 0, 1, 2, 5]
steepness_values = [1/10, 1/3, 1/2, 1, 2, 3, 10]

fig, axs = plt.subplots(2, 7, figsize=(35, 7))

for i, x0 in enumerate(midpoint_values):
    axs[0, i].plot(x, sigmoid(x, 1, x0))
    axs[0, i].set_title(f"Midpoint: {x0}")

for i, k in enumerate(steepness_values):
    axs[1, i].plot(x, sigmoid(x, k, 0))
    axs[1, i].set_title(f"Steepness: {k:.2f}")

plt.show()
```
